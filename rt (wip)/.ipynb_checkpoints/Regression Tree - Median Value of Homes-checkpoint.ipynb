## Importing Required Libraries

```python
import matplotlib.pyplot as plt
import pandas
import numpy
import seaborn as sns
import missingno as msno
%matplotlib inline
```

## Reading CSV File

```python
df = pandas.read_csv("real_estate_data.csv")
df.head()
```

## Preprocessing Data

### Checking for Missing Values

```python
df.isna().sum()
```

Since there are some missing values, we need to decide that to replace them with: the mean, the median, or the mode. Typically, if there are less outliers, we can use mean. Let's check.

```python
print(df.shape)

df[df.isnull().any(axis=1)]
```

There are a total of 506 rows, out of which 112 have missing `NaN` values. We need to deal with them by either dropping those rows (CCA; only if less than `5%` of data is randomly missing, the distribution remains the same before and after, and target value counts remain relatively the same before and after as well), forward filling, backward filling, or replacing with the mean/median/mode.

As we can see, less than `5%` data is missing in all columns (as evident fromt he table) and it is randomly distributed (as evident from the grid), so CCA is possible in all columns. However, we are left with `77.9%` data, which is slightly low.

```python
print(df.isnull().mean()*100)

cols_with_nan = []

for columns in df:
    if df[columns].isnull().mean() < 0.05 and df[columns].isnull().mean() > 0 and columns != 'MEDV':
        cols_with_nan.append(columns)

print("\nData leftover after dropping missing rows:", len(df[cols_with_nan].dropna())/len(df)*100)

msno.matrix(df)
plt.show()
```
