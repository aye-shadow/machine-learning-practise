# MLR: Predicting Final Grade

## Importing Libraries

```python
import pandas as pd
import numpy as np
import sklearn
from sklearn import linear_model
import matplotlib.pyplot as plt
```

## Reading the CSV File

```python
data = pd.read_csv("student-mat.csv", sep=";")
data.head()
```

## Formatting Categorical Data Columns

```python
for column in data:
    if data[column].dtype == object:
        print(column + ":", data[column].unique())
```

All binary-data columns can be formatted using label encoding. Categorical data can be formatted using label encoding iff it follows an order (e.g. "middle school", "high school", "college"); here, it does not follow any strict order, so we will use one hot encoding for it instead.

### Label Encoding

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

for column in data:
    if data[column].dtype == object and len(data[column].unique()) == 2:
        data[column] = encoder.fit_transform(data[column])

data.head()
```

### One Hot Encoding

```python
for column in data:
    if data[column].dtype == object:
        data = pd.concat([data, pd.get_dummies(data[column], dtype=int, prefix=column+"_")], axis=1)
        data = data.drop([column], axis=1)

data.head()
```

## Visualing Linearity

```python
label = "G3"

x = 1

plt.figure(figsize=(20, 40))

for column in data:
    if column != label:
        plt.subplot(10, 5, x)
        plt.scatter(data[column], data[label], color='blue')
        plt.xlabel(column)
        plt.ylabel("G3")

        x += 1
        
plt.show()
```

The only variables with a linear relationship with our label are `G1` and `G2`, so those are the only independent variables we will select for our linear regression model.

```python
for column in data:
    if not(column == "G1" or column == "G2" or column == label):
        data = data.drop([column], axis=1)

X = data.drop([label], axis=1)
y = data[label]
```

## Checking Multicollinearity

```python
import seaborn as sns

corr = data.corr()
print(corr)
```

This correlation matrix shows us that `G1` and `G2` are highly correlated, and since the correlation of `G2` is stronger than `G1`'s with `G3`, we can remove `G1` if we wish to reduce the multicollinearity.

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools import add_constant

X1 = add_constant(data)
series_before = pd.Series([variance_inflation_factor(X1.values, i) for i in range(X1.shape[1])], index = X1.columns)
print(series_before)
```

There is no VIF score > 10, indicating that there is no problematic multicollinearity.

## Defining Attributes and Label

```python
X = data.drop([label], axis=1)
y = data[label]
```

## Splitting Data

```python
from sklearn import model_selection 

X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size = 0.2, random_state = 0)
```

## Scaling Data

In Linear Regression, it is advised that we scale our X-train-set before training our model on it as we do not want any one feature to dominate over the others. Here, I am standardising them (making mean = standard deviation = 0)

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_train)
X_scaled = pd.DataFrame(X_train, columns=X_train.columns)
```

## Training the Model

```python
linear = linear_model.LinearRegression().fit(X_scaled, y_train)

y_pred = linear.predict(X_test)
```

### Evaluating Results

```python
plt.scatter(y_test, y_pred)
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Actual vs Predicted')
plt.show()
```

```python
from sklearn.metrics import r2_score

print(r2_score(y_test, y_pred))
```
